```python
DataPath = 'gdrive/My Drive/PhD/Thesis/Ideas/Codes/CASAS/AL-Smarthome/myData'
filePath = DataPath + '/AR_TEST'
```

```python
import pandas as pd
from datetime import datetime
import pytz
import ast
```

```python
df = pd.read_csv(filePath)

#change string to list of integer
df['motion sensors'] = df['motion sensors'].apply(ast.literal_eval)
df['motion sensors'] = df['motion sensors'].apply(lambda s: list(map(int, s)))

df['beacon sensors'] = df['beacon sensors'].apply(ast.literal_eval)
df['beacon sensors'] = df['beacon sensors'].apply(lambda s: list(map(int, s)))
```

```python
#return name of the sensor
def Name(number, typeSensor):
  if number<10:
    return typeSensor + str(0) + str(number)
  else:
    return typeSensor + str(number)

#convert epoch time to human readable
def convertTime(posix_timestamp):
  tz = pytz.timezone('MST')
  dt = datetime.fromtimestamp(posix_timestamp, tz)
  time = dt.strftime('%Y-%m-%d %H:%M:%S.%f')[:-3]
  return time


previous_M = None
previous_B = None
for index, row in df.iterrows():
  T = row['time']
  M = row['motion sensors']
  B = row['beacon sensors']
  Activity = row['activity']
  time = convertTime(T)
  MotionSensor_Names = []
  MotionSensor_Message = []
  BeaconSensor_Names = []
  BeaconSensor_Message = []
  # Motion Sensor
  for i in range(len(M)):
    if M[i] == 1:
      MotionSensor_Names.append(Name(i,'M'))
      MotionSensor_Message.append('ON')
    if previous_M != None:
      if M[i] == 0 and previous_M[i] == 1: 
        MotionSensor_Names.append(Name(i,'M'))
        MotionSensor_Message.append('OFF')
  previous_M = M
  # Beacon Sensor
  for i in range(len(B)):
    if B[i] == 1:
      BeaconSensor_Names.append(Name(i,'B'))
      BeaconSensor_Message.append('ON')
    if previous_B != None:
      if B[i] == 0 and previous_B[i] == 1: 
        BeaconSensor_Names.append(Name(i,'B'))
        BeaconSensor_Message.append('OFF')
  previous_B = B
  #print(time, MotionSensor_Names, MotionSensor_Message, BeaconSensor_Names, BeaconSensor_Message, Activity)
  #save to file
  with open(filePath + '_processed', 'a') as output_file:
    for m in range(len(MotionSensor_Names)):
      output_file.write(time +' '+ MotionSensor_Names[m] + ' ' + MotionSensor_Names[m] + ' ' + MotionSensor_Message[m] + ' ' + Activity + '\n')
    #!for b in range(len(BeaconSensor_Names)):
    #!  output_file.write(time +' '+ BeaconSensor_Names[b] + ' ' + BeaconSensor_Names[b] + ' ' +BeaconSensor_Message[b] + ' ' + Activity + '\n')   
    
```
